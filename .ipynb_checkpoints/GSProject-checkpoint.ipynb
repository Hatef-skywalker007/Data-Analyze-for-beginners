```python
sale_platform1 = data_two.groupby(["Year","Platform"])["Global_Sales"].sum().reset_index()
sale_platform1
```

```python
#region
plt.figure(figsize=(12,6))
sb.lineplot(data=sale_platform1, x="Year" , y = "Global_Sales",hue="Platform",marker="o")
plt.title("Sale per Year by platform")
plt.ylabel("Total sales",rotation=90)
plt.grid()
plt.show
```

```python
sale_year1 = data_two.groupby("Year")[["NA_Sales","EU_Sales","JP_Sales","Other_Sales"]].sum().reset_index()
sale_m = sale_year1.melt(id_vars = "Year" , var_name = "Region",value_name = "Sales")#ادغام کردن
plt.figure(figsize=(10,6))
sb.lineplot(data = sale_m,x = "Year",y = "Sales",hue ="Region",marker = "o")
plt.title("Region Sales over time")
plt.ylabel("Sales",rotation = 90)
plt.grid()
plt.show()
```

```python
#genre
genre_sale = data_two.groupby("Genre")["Global_Sales"].sum().sort_values().reset_index()
genre_sale
```

```python
plt.figure(figsize = (10,6))
sb.barplot(data = genre_sale , x = "Global_Sales" , y = "Genre" , palette = "dark" , hue = "Genre")
plt.title("Total Sales by Genre")
plt.ylabel("Genre",rotation = 90)
plt.xlabel("Total Sales")
plt.grid()
plt.show()
```

```python
sale_genre = data_two.groupby("Genre")[["NA_Sales","EU_Sales","JP_Sales","Other_Sales"]].sum().reset_index()
sale_m1 = sale_genre.melt(id_vars = "Genre" , var_name = "Region",value_name = "Sales")
sale_m1
```
